```python
import rasterio
import numpy as np
from sklearn.preprocessing import minmax_scale
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt

red_file = rasterio.open('/home/jovyan/work/ESDA_xjw/Project/data/modis_20210914/red.tif')
red_data = red_file.read(1)
red_data[red_data < 0] = 0

green_file = rasterio.open('/home/jovyan/work/ESDA_xjw/Project/data/modis_20210914/green.tif')
green_data = green_file.read(1)
green_data[green_data < 0] = 0

blue_file = rasterio.open('/home/jovyan/work/ESDA_xjw/Project/data/modis_20210914/blue.tif')
blue_data = blue_file.read(1)
blue_data[blue_data < 0] = 0

nir_file = rasterio.open('/home/jovyan/work/ESDA_xjw/Project/data/modis_20210914/nir.tif')
nir_data = nir_file.read(1)
nir_data[nir_data < 0] = 0

ir_file = rasterio.open('/home/jovyan/work/ESDA_xjw/Project/data/modis_20210914/ir.tif')
ir_data = ir_file.read(1)
ir_data[ir_data < 0] = 0

fir_file = rasterio.open('/home/jovyan/work/ESDA_xjw/Project/data/modis_20210914/fir.tif')
fir_data = fir_file.read(1)
fir_data[fir_data < 0] = 0

ffir_file = rasterio.open('/home/jovyan/work/ESDA_xjw/Project/data/modis_20210914/ffir.tif')
ffir_data = ffir_file.read(1)
ffir_data[ffir_data < 0] = 0

glims_file = '/home/jovyan/work/ESDA_xjw/Project/data/iceland_glims/glims_polygons.tif'
glims_data = rasterio.open(glims_file).read(1)
glims_data = np.where(glims_data > 0, 1, 0)

# x is input, y is output
y = glims_data
# converting a 2D list into a 1D list
y_reshaped = y.flatten()
X = np.dstack((red_data, nir_data, green_data, blue_data, ir_data, fir_data, ffir_data))
X_reshaped = X.reshape(len(y_reshaped), X.shape[2])

X_test = X_reshaped[0:len(y_reshaped)//2]
X_train = X_reshaped[len(y_reshaped)//2:len(y_reshaped)]
y_test = y_reshaped[0:len(y_reshaped)//2]
y_train = y_reshaped[len(y_reshaped)//2:len(y_reshaped)]

knn = KNeighborsClassifier(n_neighbors=6)
knn.fit(X_train, y_train)
print(knn.score(X_test, y_test))
output_flatten = knn.predict(X_reshaped)
output = output_flatten.reshape(X.shape[0], X.shape[1])
plt.figure(figsize=(20, 20))
plt.imshow(output)

square_area = (18.3 - 14.7) * 111 * (65 - 63.7) * 111
area_per_pixel = square_area / len(y_reshaped)
glacier_area21 = np.sum(output) * area_per_pixel
print(glacier_area21)
```

```python
import numpy as np
import matplotlib.pyplot as plt

glacier_areas = [glacier_area00, glacier_area01, glacier_area02, 
                 glacier_area04, glacier_area06, glacier_area08, 
                 glacier_area10, glacier_area14,
                 glacier_area17, glacier_area21]

years = [2000, 2001, 2002, 2004, 2006, 2008, 2010, 2014, 2017, 2021]

y = np.floor(glacier_areas)
x = np.array(years)

b, m = np.polynomial.polynomial.polyfit(x, y, 1)

plt.figure()
plt.plot(x, y, '.')
plt.plot(x, b + m * x, '-')
plt.xlabel("Year")
plt.ylabel("Area (meter square)")
plt.ylim([17500, 22000])
plt.title("Glacier Change over from 2000 to 2020",fontsize=15)
annotations = x
for i, label in enumerate(annotations):
    plt.annotate(label, (x[i], y[i]))

plt.savefig("fig.png")
```
